## Apply function

```python
df.head()
```

```python
def preg_func(p):
    if(p<=2):
        return "Normal"
    else:
        return "High"
```

```python
df['status']=df['Pregnancies'].apply(preg_func)
```

Lambda Function

```python
df['status1']=df['Pregnancies'].apply(lambda x:"Normal" if x<=2 else "High")
```

## Groupby Functions

```python
df_grp_preg=df.groupby('status')
```

```python
df_grp_preg.mean()
```
