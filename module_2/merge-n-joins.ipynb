```python
import pandas as pd
import numpy as np
```

## Normal Concatenation


```python
data1 = {'Name':['Akash', 'Ron', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd'],
        'Mobile No': [97, 91, 58, 76]} 
   
# Define a dictionary containing employee data 
data2 = {'Name':['Gaurav', 'Anuj', 'Harry', 'Ela'], 
        'Age':[22, 32, 12, 52], 
        'Address':['Allahabad', 'Kannuaj', 'Allahabad', 'Kannuaj'], 
        'Qualification':['MCA', 'Phd', 'Bcom', 'B.hons'],
        'Salary':[1000, 2000, 3000, 4000]} 
 
# Convert the dictionary into DataFrame  
df = pd.DataFrame(data1,index=[0, 1, 2, 3])
 
# Convert the dictionary into DataFrame  
df1 = pd.DataFrame(data2, index=[2, 3, 6, 7]) 
 
print(df, "\n\n", df1) 
```

```python
con=pd.concat([df,df1])
con
```

```python
con1=pd.concat([df,df1],axis=1,sort=True)
con1
```

```python
con2=pd.concat([df,df1],axis=1,join="inner")
con2
```

## Merge

```python
pd.merge(df,df1,on=['Name','Age'])
```

```python
data1 = {'key': ['K0', 'K1', 'K2', 'K3','K5'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj','Raj'], 
        'Age':[27, 24, 22, 32,67],} 
   
# Define a dictionary containing employee data 
data2 = {'key': ['K0', 'K1', 'K2', 'K3','k6'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj','Kolkata'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons','Btech']} 
 
# Convert the dictionary into DataFrame  
df = pd.DataFrame(data1)
 
# Convert the dictionary into DataFrame  
df1 = pd.DataFrame(data2) 
  
 
print(df, "\n\n", df1) 
```

```python
pd.merge(df,df1,on='key')
```

```python
pd.merge(df,df1,on='key',how='outer')
```

```python
pd.merge(df,df1,on='key',how='left')
```

```python
pd.merge(df,df1,on='key',how='right')
```
